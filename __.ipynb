```python
def get_files(filepath):
    all_files = []
    for root, dris, files in os.walk(filepath):
        files = glob.glob(os.path.join(root,'*.json'))
        for f in files:
            all_files.append(os.path.abspath(f))

    return all_files
```

```python
import pandas as pd

def get_over_data():
    innings = exp['innings']
    deliveries_list = []
    
    for inning_data in innings:
        inning_name = inning_data.get('team', 'Unknown Team')  # Get team name for context
        overs = inning_data.get('overs', [])
        
        for over_data in overs:
            over_number = over_data.get('over', None)  # Extract over number
            deliveries = over_data.get('deliveries', [])
            
            for delivery in deliveries:
                delivery['inning'] = inning_name
                delivery['over'] = over_number
                
                # Extract run details if available
                runs = delivery.get('runs', {})
                delivery['runs_batter'] = runs.get('batter', 0)
                delivery['runs_extras'] = runs.get('extras', 0)
                delivery['runs_total'] = runs.get('total', 0)
                
                deliveries_list.append(delivery)  # Add each delivery as a row

    # Convert to DataFrame and drop the original 'runs' column
    df = pd.DataFrame(deliveries_list).drop(columns=['runs'], errors='ignore')

    return df  

df = get_over_data()
print(df.head())  # Display first few rows
```
